```python
# Importing required libraries -----------------
import numpy as np
import matplotlib.pyplot as plt
import time
import torch
from torch import nn
from torch import optim
from torch.nn import functional as F
from torchvision import transforms
from torchvision import models
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

import pandas as pd
from coordinates import first,second,third,forth,exti 
from PIL import Image
# import cv2
import statistics
from PIL import Image
# from torch.autograd import Variable
from IPython.display import Image

%matplotlib inline

import traitlets
from ipywidgets import widgets
from IPython.display import display
from tkinter import Tk, filedialog

from IPython.display import Markdown, display

# User defined Classes --------------------------

# Select file button class - To input the image file by browsing from SelectFile button

class SelectFilesButton(widgets.Button):
    def __init__(self):
        super(SelectFilesButton, self).__init__()
        self.add_traits(files=traitlets.traitlets.List())
        self.description = "Select Image"
        self.icon = "square-o"
        self.style.button_color = "orange"

        self.on_click(self.select_files)

    @staticmethod
    def select_files(b):
        with out:
            try:
                root = Tk()
                root.withdraw()
                root.call('wm', 'attributes', '.', '-topmost', True)
                b.files = filedialog.askopenfilename(multiple=True)
                b.description = "Image Selected"
                b.icon = "check-square-o"
                b.style.button_color = "lightgreen"
            except:
                pass

            
## Model Class - We are defining the model here

class Pdf(nn.Module):
    def __init__(self):
        super(Pdf, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, stride=2, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, stride=2, padding=1)
        self.conv3 = nn.Conv2d(64, 128, 3, padding=1)
        self.pool = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(7*7*128, 500)
        self.fc2 = nn.Linear(500,6)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 7*7*128)
        x = F.relu(self.fc1(self.dropout(x)))
        x = self.fc2(self.dropout(x))
        return x

## Convolutional neural network model is designed as below
    """
    (conv1): Conv2d(3, 32, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1))
    (conv2): Conv2d(32, 64, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1))
    (conv3): Conv2d(64, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1))
    (pool): MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)
    (fc1): Linear(in_features=6272, out_features=500, bias=True)
    (fc2): Linear(in_features=500, out_features=6, bias=True)
    (dropout): Dropout(p=0.5)    
    
    """
    

# User defined functions -------------

def values_normal(x):
    """
    returns a value 
    for x ie if x is 2 , it will return 212
    """
    if x==2:
        return 212
    elif x==3:
        return 213
    elif x==4:
        return 214
    elif x==5:
        return 215
    elif x==1:
        return 211
    elif x==0:
        return 210
    
    
def create_download_link( df, title = "Download CSV file", filename = "Mold_Output.csv"):
    """
    it creates download link for csv file
    """
    from IPython.display import HTML
    import pandas as pd
    import numpy as np
    import base64
    csv = df.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)


def printmd(string):
    """
    Customizing print statements
    """
    display(Markdown(string))

def print_cuda_availability():     
    if use_cuda : 
        print("This system has CUDA. Program will be running in GPU")#gpu
    else :
        print("This system does not have CUDA. So the program will be running in CPU")#cpu

        
## Image preprocessing - Gets outputs and corresponding probabilities

def get_output_and_probabilities(input_file):   
    size=(224,224)
    moldo=[]
    moldp=[]
    img = cv2.imread(input_file,0)
    plt.imshow(img,cmap='gray')
    plt.title('input')
    plt.show()
    a=img.flatten()
    thresh=(statistics.mean(a)-((a.std()/2)+5))
    (thresh, img_bin) = cv2.threshold(img, thresh, 255,cv2.THRESH_BINARY|cv2.THRESH_OTSU)
    img_bin = 255-img_bin
    cv2.imwrite("outputs/invert.jpg",img_bin)
    plt.imshow(img_bin,cmap='gray')
    plt.title('invert')
    plt.show()
    top, contours = cv2.findContours(img_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for c in top:
        epsilon = 0.008*cv2.arcLength(c,True)
        approx = cv2.approxPolyDP(c,epsilon,True)
        x, y, w, h = cv2.boundingRect(approx)
        if w>2000 and (h>50 and h<100):
            cropt = img[y:y+1600,x:x+w]
            cropt = cv2.resize(cropt,(2118,1600))
            cv2.imwrite("outputs/resized.jpg",cropt)
            plt.imshow(cropt,cmap='gray')
            plt.title('resized image')
            plt.show()
    imi = cropt
    idx=0
    for c in exti:
            x, y, w, h = c
            idx+=1
            if idx==4:
                f=0
                crop_img = imi[y:y+h, x:x+w]
                cv2.imwrite("outputs/cropped/4.jpg",crop_img)
                plt.imshow(crop_img,cmap='gray')
                plt.title('outer part crop')
                plt.show()
                for value in forth:
                    f+=1
                    x, y, w, h = value
                    cv2.rectangle(crop_img,(x,y),(x+w,y+h),(0,0,0),2)
                    crop = crop_img[y:y+h, x:x+w]
                    cv2.imwrite("outputs/4/crop"+str(f)+".jpg",crop)
                    if f>=2 and f<=7:
                        plt.imshow(crop,cmap='gray')
                        plt.title('inner part crop')
                        plt.show()
                    image_transform=transforms.Compose([transforms.ToPILImage(),
                                    transforms.Resize(size, interpolation=2),
                                        transforms.Grayscale(3),
                                     transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                              std=[0.229, 0.224, 0.225])])
                    image = image_transform(crop)
                    image = image.unsqueeze(0)
                    data = image.float()
                    if use_cuda:
                        data = data.cuda()
                    output = scratch(data)
                    pred = output.data.max(1, keepdim=True)[1]
                    sm = torch.nn.Softmax(dim=1)
                    output1 = sm(output) 
                    prob,label = torch.topk(output1,1)
                    output = label.data.cpu().numpy()[0][0]
                    probs = prob.data.cpu().numpy()[0][0]
                    moldo.append(output)
                    moldp.append(probs)
            elif idx==3:
                t=0
                crop_img = imi[y:y+h, x:x+w]
                cv2.imwrite("outputs/cropped/3.jpg",crop_img)
                for value in third:
                    t+=1
                    x, y, w, h = value
                    cv2.rectangle(crop_img,(x,y),(x+w,y+h),(0,0,0),2)
                    crop = crop_img[y:y+h, x:x+w]
                    cv2.imwrite("outputs/3/crop"+str(t)+".jpg",crop)
                    image_transform=transforms.Compose([transforms.ToPILImage(),
                                    transforms.Resize(size, interpolation=2),
                                        transforms.Grayscale(3),
                                     transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                              std=[0.229, 0.224, 0.225])])
                    image = image_transform(crop)
                    image = image.unsqueeze(0)
                    data = image.float()
                    if use_cuda:
                        data = data.cuda()
                    output = scratch(data)
                    pred = output.data.max(1, keepdim=True)[1]
                    sm = torch.nn.Softmax(dim=1)
                    output1 = sm(output) 
                    prob,label = torch.topk(output1,1)
                    output = label.data.cpu().numpy()[0][0]
                    probs = prob.data.cpu().numpy()[0][0]
                    moldo.append(output)
                    moldp.append(probs)
            elif idx==2:
                s=0
                crop_img = imi[y:y+h, x:x+w]
                cv2.imwrite("outputs/cropped/2.jpg",crop_img)
                for value in second:
                    s+=1
                    x, y, w, h = value
                    cv2.rectangle(crop_img,(x,y),(x+w,y+h),(0,0,0),2)
                    crop = crop_img[y:y+h, x:x+w]
                    cv2.imwrite("outputs/2/crop"+str(s)+".jpg",crop)
                    image_transform=transforms.Compose([transforms.ToPILImage(),
                                    transforms.Resize(size, interpolation=2),
                                        transforms.Grayscale(3),
                                     transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                              std=[0.229, 0.224, 0.225])])
                    image = image_transform(crop)
                    image = image.unsqueeze(0)
                    data = image.float()
                    if use_cuda:
                        data = data.cuda()
                    output = scratch(data)
                    pred = output.data.max(1, keepdim=True)[1]
                    sm = torch.nn.Softmax(dim=1)
                    output1 = sm(output) 
                    prob,label = torch.topk(output1,1)
                    output = label.data.cpu().numpy()[0][0]
                    probs = prob.data.cpu().numpy()[0][0]
                    moldo.append(output)
                    moldp.append(probs)
            elif idx==1:
                o=0
                crop_img = imi[y:y+h, x:x+w]
                cv2.imwrite("outputs/cropped/1.jpg",crop_img)
                for value in first:
                    o+=1
                    x, y, w, h = value
                    cv2.rectangle(crop_img,(x,y),(x+w,y+h),(0,0,0),2)
                    crop = crop_img[y:y+h, x:x+w]
                    cv2.imwrite("outputs/1/crop"+str(o)+".jpg",crop)
                    image_transform=transforms.Compose([transforms.ToPILImage(),
                                    transforms.Resize(size, interpolation=2),
                                        transforms.Grayscale(3),
                                     transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                              std=[0.229, 0.224, 0.225])])
                    image = image_transform(crop)
                    image = image.unsqueeze(0)
                    data = image.float()
                    if use_cuda:
                        data = data.cuda()
                    output = scratch(data)
                    pred = output.data.max(1, keepdim=True)[1]
                    sm = torch.nn.Softmax(dim=1)
                    output1 = sm(output) 
                    prob,label = torch.topk(output1,1)
                    output = label.data.cpu().numpy()[0][0]
                    probs = prob.data.cpu().numpy()[0][0]
                    moldo.append(output)
                    moldp.append(probs)
                    
    return moldo, moldp

## Variable Assignments --------   

# Storing availability of cuda
use_cuda = torch.cuda.is_available()

# Storing the model
if use_cuda:
    scratch = Pdf().cuda()
else :
    scratch = Pdf()

# Selecting and Storing Image    
out = widgets.Output()
raw = SelectFilesButton()
#widgets.VBox([raw, out])


```

### 4. Image Preprocessing
- The following cell may take 10 seconds to run
- Below are the various steps followed while processing the image:
    1. Resizing images for better quality 
    2. Inverting images for better threshold
    3. Finding the contours from the inverted images
    4. Finding the coordinates of the repective contours
    5. Cropping of templates
    6. Scaling the cropped images
    7. Passing the cropped images to the CNN model
    8. Appending the values to list

```python
moldo, moldp = get_output_and_probabilities(input_file)
```

### 5.Output
- Please run the cell below and 
- Dowload the output using the 'Download CSV file' link that will appear below this cell

```python
molds = pd.DataFrame()
tl,tlp,tr,trp,tm,tmp,bl,blp,br,brp,bm,bmp,date = [],[],[],[],[],[],[],[],[],[],[],[],[]
tl1,tlp1,tr1,trp1,tm1,tmp1,bl1,blp1,br1,brp1,bm1,bmp1 = [],[],[],[],[],[],[],[],[],[],[],[]
idx = 0


for (output,probs) in zip (moldo,moldp):
        idx+=1
        if (idx>=33 and idx<=48):
            tl.append(output)
            tlp.append(probs)
        elif(idx>=175 and idx<=189):
            tl1.append(output)
            tlp1.append(probs)
        if (idx>=49 and idx<=64):
            tm.append(output)
            tmp.append(probs)
        elif(idx>=190 and idx<=204):
            tm1.append(output)
            tmp1.append(probs)
        if (idx>=65 and idx<=80):
            tr.append(output)
            trp.append(probs)
        elif(idx>=205 and idx<=219):
            tr1.append(output)
            trp1.append(probs)
        if (idx>=81 and idx<=96):
            bl.append(output)
            blp.append(probs)
        elif(idx>=220 and idx<=234):
            bl1.append(output)
            blp1.append(probs)
        if (idx>=97 and idx<=112):
            bm.append(output)
            bmp.append(probs)
        elif(idx>=235 and idx<=249):
            bm1.append(output)
            bmp1.append(probs)
        if (idx>=113 and idx<=128):
            br.append(output)
            brp.append(probs)
        elif(idx>=250 and idx<=264):
            br1.append(output)
            brp1.append(probs)
    
tl.reverse()
tl1.reverse()
tl =tl + tl1
tm.reverse()
tm1.reverse()
tm =tm+tm1
tr.reverse()
tr1.reverse()
tr =tr+tr1
bl.reverse()
bl1.reverse()
bl = bl+bl1
bm.reverse()
bm1.reverse()
bm =bm+bm1
br.reverse()
br1.reverse()
br =br+br1
tlp.reverse()
tlp1.reverse()
tlp =tlp+tlp1
tmp.reverse()
tmp1.reverse()
tmp =tmp1+tmp
trp.reverse()
trp1.reverse()
trp = trp+trp1
blp.reverse()
blp1.reverse()
blp =blp+blp1
bmp.reverse()
bmp1.reverse()
bmp =bmp+bmp1
brp.reverse()
brp1.reverse()
brp =brp+brp1

for i in range(1,len(tl)+1):
    date.append(str(i)+"-10-2017")

    
molds['date'] = date
molds['tl'] = tl
molds['tm'] = tm
molds['tr'] = tr
molds['bl'] = bl
molds['bm'] = bm
molds['br'] = br
molds['tlp'] = tlp
molds['tmp'] = tmp
molds['trp'] = trp
molds['blp'] = blp
molds['bmp'] = bmp
molds['brp'] = brp

threshold = 0.88

molds.loc[(molds['tlp'] > threshold),'tlf'] = 'Certain'
molds.loc[(molds['tlp']<=threshold),'tlf'] = 'Uncertain'
molds.loc[(molds['tmp'] > threshold),'tmf'] = 'Certain'
molds.loc[(molds['tmp']<=threshold),'tmf'] = 'Uncertain'
molds.loc[(molds['trp'] > threshold),'trf'] = 'Certain'
molds.loc[(molds['trp']<=threshold),'trf'] = 'Uncertain'
molds.loc[(molds['blp'] > threshold),'blf'] = 'Certain'
molds.loc[(molds['blp']<=threshold),'blf'] = 'Uncertain'
molds.loc[(molds['bmp'] > threshold),'bmf'] = 'Certain'
molds.loc[(molds['bmp']<=threshold),'bmf'] = 'Uncertain'
molds.loc[(molds['brp'] > threshold),'brf'] = 'Certain'
molds.loc[(molds['brp']<=threshold),'brf'] = 'Uncertain'
molds[['tl','tm','tr','bl','bm','br']]=molds[['tl','tm','tr','bl','bm','br']].applymap(lambda x: values_normal(x))
create_download_link(molds)
```
